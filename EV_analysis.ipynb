```python
states_2W['Penetration Rate'] = (states_2W['electric_vehicles_sold']/states_2W['total_vehicles_sold'])*100
states_2W.head(369)
```

```python
states_2W['date_parsed'] = pd.to_datetime(states_2W['date'], format='%d-%b-%y')
```

```python
states_2W.head(10)
```

```python
states_2W['FY'] = states_2W['date_parsed'].dt.year + (states_2W['date_parsed'].dt.month>=4).astype(int)
states_2W.head(10)
```

```python
agg = (
    states_2W
    .groupby(['state', 'FY'])
    .agg(
      ev_sold    = ('electric_vehicles_sold', 'sum'),
      total_sold = ('total_vehicles_sold',   'sum')
    )
    .reset_index()
)
```

```python
agg.head(4)
```

```python
agg['penetration'] = agg['ev_sold'] / agg['total_sold']
```

```python
prev = agg[['state', 'FY', 'penetration']].copy()
```

```python
prev.head(4)
```

```python
prev['FY'] += 1
prev.rename(columns={'penetration':'pen_prev'}, inplace=True)
```

```python
prev.head(4)
```

```python
yoy = agg.merge(prev, on=['state','FY'], how='left')
```

```python
yoy.head(100)
```

```python
yoy['yoy_change'] = yoy['penetration'] - yoy['pen_prev']
```

```python
yoy_neg = yoy[
    (yoy['FY'].isin([2022, 2023, 2024]))
    & (yoy['yoy_change'] < 0)
]
```

```python
yoy_neg
```

```python
 for fy in [2022, 2023, 2024]:
    print(yoy_neg[yoy_neg['FY'] == fy])
```

```python
results = []
for fy in [2022, 2023, 2024]:
    subset = yoy_neg[yoy_neg['FY'] == fy]
    # print(subset)
    if subset.empty:
        results.append({'FY': fy, 'most_negative_states': None, 'least_negative_states': None})
        continue

    most_drop  = subset['yoy_change'].min()
    least_drop = subset['yoy_change'].max()

    most_states  = subset[subset['yoy_change'] == most_drop ]['state'].tolist()
    least_states = subset[subset['yoy_change'] == least_drop]['state'].tolist()

    results.append({
      'FY': fy,
      'most_negative_states':  most_states,
      'least_negative_states': least_states
    })
```

```python
results
```

```python
res_df = pd.DataFrame(results)
print(res_df)
```

```python
res_df.to_csv('NegStates.csv',index=False)
```
